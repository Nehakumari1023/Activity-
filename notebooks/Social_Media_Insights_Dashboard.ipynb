*Interest Analysis*

```python
from collections import Counter

all_interests = []

for row in df1['Interests'].dropna():
    all_interests.extend(
        [x.strip() for x in row.split(',')]
    )

Counter(all_interests).most_common(20)
```

```python
interest_counts = Counter(all_interests).most_common(20)

interest_df = pd.DataFrame(
    interest_counts,
    columns=['Interest', 'Count']
)

plt.figure(figsize=(10,8))

sns.barplot(
    data=interest_df,
    x='Count',
    y='Interest',
    palette='viridis'
)

plt.title('Top 20 User Interests')
plt.xlabel('Number of Users')
plt.ylabel('Interest')

plt.show()
```

The largest user segment belongs to the 19–24 age group.

```python
df1['AgeGroup'] = pd.cut(
    df1['Age'],
    bins=[13,18,24,30,40,60],
    labels=[
        '13-18',
        '19-24',
        '25-30',
        '31-40',
        '41-60'
    ]
)
```

```python
df1['AgeGroup'].value_counts()
```

```python
age_counts = df1['AgeGroup'].value_counts().sort_index()

plt.figure(figsize=(10,6))

colors = [
    '#FFB6C1',  # 13-18
    '#87CEFA',  # 19-24
    '#98FB98',  # 25-30
    '#FFD700',  # 31-40
    '#FFA07A'   # 41-60
]

ax = sns.barplot(
    x=age_counts.index,
    y=age_counts.values,
    palette=colors
)

plt.title(
    'Social Media Users by Age Group',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Age Group')
plt.ylabel('Number of Users')

# Add values on bars
for i, v in enumerate(age_counts.values):
    plt.text(i, v + 500, f'{v:,}', ha='center')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

The 41–60 age group represents the largest segment of social media users, followed by users aged 31–40 and 25–30. The 13–18 age group is absent from the dataset.
